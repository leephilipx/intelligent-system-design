```python
import cv2
import numpy as np
```

```python
import glob
import cv2
import os

images = glob.glob("retina2/*.bmp")
```

```python
dstpath = 'Cleaned' # Destination Folder
minLineLength = 30
maxLineGap = 5
```

```python
for path in images:
    img = cv2.imread(path)
#     gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    edges = cv2.Canny(img, 80, 255)
    lines = cv2.HoughLinesP(edges,cv2.HOUGH_PROBABILISTIC, np.pi/180, 30, minLineLength,maxLineGap)
    for line in lines:
        x1, y1, x2, y2 = line[0]
        cv2.line(img, (x1, y1), (x2, y2), (0,128,0),2, cv2.LINE_AA)
    out = dstpath + '/' + os.path.basename(path).replace('.bmp', '.png')
    print(img.shape)
    cv2.imwrite(out, img)
    
```
